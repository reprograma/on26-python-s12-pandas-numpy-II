## Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Carregando os dados

```python
# carregando os dados
df = pd.read_csv('titanic.csv')
df
```

## Mudando o display da tela

```python
# Option
display_settings = {
    'max_columns': 999,
    'width': 4096,    
    'max_rows': 999,
    'precision': 2,
}

for op, value in display_settings.items():
    pd.set_option("display.{}".format(op), value)
```

```python
df.head(100)
```

## Gráficos

```python
df['Age'].plot.hist()  
```

```python
# Gráfico númerico (1 variável)
df['Age'].plot.hist(bins=20, edgecolor='black', color="orange")  

plt.xlabel('Classe', fontweight='bold') 
plt.ylabel('Quantidade') 
plt.title('Distribuição de Classes', color='darkred')

plt.tick_params(axis='x', colors='red')
plt.tick_params(axis='y', colors='purple')
```

```python
# análise descritiva da var 'Age'
df[['Age']].describe(percentiles=[0.005, 0.10, 0.80, 0.995]).T
```

```python
df.loc[(df['Age'] == 0.42), :]
```

```python
df.loc[(df['Age'] == 80), :]
```

```python
# Gráfico númerico (1 variável)

# filtro com condicional
filtro_hist = (df['Age'] > 0.8) & (df['Age'] < 70.72)

# gerando o histograma
df.loc[filtro_hist, 'Age'].plot.hist(bins=20, edgecolor='black')

plt.xlabel('Classe') 
plt.ylabel('Quantidade') 
plt.title('Distribuição de Classes')
```
