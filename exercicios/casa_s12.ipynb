```python
# Exercício Casa - Semana 12
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('netflix.csv')
df
```

```python
# Tradução das colunas

traducoes = {
    'User ID': 'Usuario ID',
    'Subscription Type': 'Tipo de Conta', 
    'Monthly Revenue': 'Receita Mensal', 
    'Join Date': 'Data Adesão',
    'Last Payment Date': 'Data Último Pagamento',
    'Country': 'País',
    'Age': 'Idade',
    'Gender':  'Gênero',
    'Device': 'Dispositivo',
    'Plan Duration': 'Duração do Plano'
    
}

novas_colunas = []
for chave, valor in traducoes.items():
    novas_colunas.append(valor)

print(novas_colunas)
```

```python
# atribuindo os novos nomes das colunas
df.columns = novas_colunas
df
```

```python
# Substituindo os valores contidos na coluna 'Subscription Type' por 'Básico' 'Comum' e 'Ouro'
df['Tipo de Conta'] = df['Tipo de Conta'].replace({'Basic': 'Básico', 'Premium': 'Ouro', 'Standard': 'Comum'})

df.head(10) # alterando o número de entradas no retorno da função
```

```python
# Substituindo os valores contidos na coluna 'Gênero' por 'Mulher' ou 'Homem'
df['Gênero'] = df['Gênero'].replace({'Male': 'Homem', 'Female': 'Mulher'})

df.head(10) # alterando o número de entradas no retorno da função
```

```python
df.shape
```

```python
# Quantas mulheres?
df_filtro_mulheres = df[df['Gênero'] == 'Mulher']
df_filtro_mulheres.shape
```

```python
# Quantas mulheres brasileiras?
df_filtro_mulheresbr = df[(df['Gênero'] == 'Mulher') & (df['País'] == 'Brazil')]
df_filtro_mulheresbr.shape
```

```python
# Mulheres brasileiras mais de 30 anos com conta básica
df_filtro_mulheres30 = df[(df['Idade'] > 30) & (df['Gênero'] == 'Mulher') & (df['Tipo de Conta'] == 'Básico') & (df['País'] == 'Brazil')]
df_filtro_mulheres30.shape
```

```python
# Dados estatísticos das usuárias brasileiras
df_filtro_mulheresbr.describe()
```

```python
# Agrupando usuários por tipo de conta
agrupado_conta = df.groupby('Tipo de Conta')

# Contagem de passageiros em cada tipo de conta, usando os dados agrupados no passo anterior
contagem_por_conta = agrupado_conta['Usuario ID'].count()

print("Contagem de usuários por tipo de conta")
print(contagem_por_conta)
```

```python
# Obter o valor máximo e mínimo da idade por usuário em cada tipo de conta:
valor_max = agrupado_conta['Idade'].max()
valor_min = agrupado_conta['Idade'].min()

print("Valor máximo da idade por tipo de conta")
print(valor_max)

print("Valor mínimo da idade por tipo de conta")
print(valor_min)
```

```python

contagem_por_conta = df['País'].value_counts()
contagem_por_conta.plot(kind='bar', color='#A08F9D');

plt.xlabel('País')
plt.title('Número de Usuários por País');
```

```python
contagem_por_conta = df['Tipo de Conta'].value_counts()
c = ['#A08F9D', '#A37E9D', '#A63B95', '#BB0FA0']
contagem_por_conta.plot(kind='bar', color=c);


plt.title('Número de Usuários por Tipo de Conta');
```

```python
contagem_por_conta = df['Gênero'].value_counts()
c = ['#A08F9D', '#A37E9D', '#A63B95', '#BB0FA0']
contagem_por_conta.plot(kind='pie', colors=c, startangle=90, autopct='%1.1f%%');


plt.title('Número de Usuários por Gênero');
```

```python
# Agrupando usuárias mulheres brasileiras por tipo de conta
agrupado_mulheresbr = df_filtro_mulheresbr.groupby('Tipo de Conta')

# Contagem de usuárias mulheres brasileiras em cada tipo de conta, usando os dados agrupados no passo anterior
contagem_por_mulheresbr = agrupado_mulheresbr['Usuario ID'].count()

print("Contagem de usuários por tipo de conta")
print(contagem_por_mulheresbr)
```
