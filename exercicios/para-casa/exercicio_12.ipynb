```python
def processamento_df(df):
    
    df_3_linhas = df.head(3)
    
    df_menor = df.sample(100)

    df = df.drop(columns='Cabin')

    df = df.sort_values("Age")

    df = df.drop_duplicates()

    df = df.reset_index()

    colunas = list(df.columns) 

    metricas = df.describe(include='all')
    
    return df_3_linhas, df_menor, df, colunas, metricas
```

```python
df_3_linhas, df_menor, df_teste, colunas, metricas = processamento_df(df)
```
