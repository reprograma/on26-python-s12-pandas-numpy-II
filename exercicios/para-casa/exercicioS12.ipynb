```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
```

# Dados do Titanic: 

Aqui iremos ver algumas informações do navio Titanic, como: 
- PassengerId (id do passageiro)
- Survived (sobreviventes)
- Pclass (classe dos passageiros)
- Name (nome)
- Sex (gênero)
- Age (idade)
- SibSp (número de irmãos)
- Parch (número de pais)
- Ticket (bilhete)
- Fare (preço do bilhete)
- Cabin (número da cabine)
- Embarked (portão de embarque)

```python
def exibir_contagem_em_grafico(data, labels, colors):
    plt.figure(figsize=(10, 6))
    bars = plt.bar(labels, data, color=colors)
    plt.title(f'Contagem')
    plt.ylabel('Quantidade')
    plt.xticks(rotation=0)
    for bar, count in zip(bars, data):
        plt.text(bar.get_x() + bar.get_width() / 2, count, str(count), ha='center', va='bottom', fontsize=12)
    plt.show()
```

### Gráfico 1: número de idosos, +60, que não sobreviveram. 

```python
idosos_sobreviventes = df[(df['Age'] > 60) & (df['Survived'] == 1)]
idosos_nao_sobreviventes = df[(df['Age'] > 60) & (df['Survived'] == 0)]

idosos_sobreviventes = df[(df['Age'] > 60) & (df['Survived'] == 1)]

idosos_nao_sobreviventes = df[(df['Age'] > 60) & (df['Survived'] == 0)]

exibir_contagem_em_grafico([len(idosos_sobreviventes), len(idosos_nao_sobreviventes)], ['Idosos Sobreviventes', 'Idosos Não Sobreviventes'], ['lightgreen', 'tomato'])
```

### Gráfico 2: idosos, +60, por classe: 

```python
idosos_por_classe = df[df['Age'] > 60]['Pclass'].value_counts()
idosos_por_classe = idosos_por_classe.sort_index()
exibir_contagem_em_grafico(idosos_por_classe.values, idosos_por_classe.index, 'skyblue')
```

### Gráfico 3: crianças, até 12, que não sobreviveram: 

```python
criancas_sobreviventes = df[(df['Age'] < 12) & (df['Survived'] == 1)]
criancas_nao_sobreviventes = df[(df['Age'] < 12) & (df['Survived'] == 0)]
exibir_contagem_em_grafico([len(criancas_sobreviventes), len(criancas_nao_sobreviventes)], ['Crianças Sobreviventes', 'Crianças Não Sobreviventes'], ['lightgreen', 'lightcoral'])
```
