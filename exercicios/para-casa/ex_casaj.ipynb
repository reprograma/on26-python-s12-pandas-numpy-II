Um breve cruzamento de dados, analisando a lista dos passageiros do titanic.

```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")
df
```

```python

#sobreviventes x porto 
sobreviventes_por_porto = df.groupby('Embarked')['Survived'].value_counts()
sobreviventes_por_porto.plot.barh()


plt.xlabel('Embarked')
plt.ylabel('Survived')
plt.title('Taxa de Sobrevivência por Porto')
```

```python
criancas_por_porto = df['Embarked'].value_counts()
criancas_por_porto.plot.bar()

plt.xlabel('Embarked')
plt.ylabel('Quantidade')
plt.title('Número de pessoas por porto')
```

```python
#quantidade de sobreviventes menores de 15 anos por clasee
sobreviventes_por_porto = df[df['Age'] < 15].groupby(['Age', 'Pclass']).size().unstack().plot.bar(stacked=True)

# Set axis labels and title
plt.xlabel('Age')
plt.ylabel('Count of Passengers')
plt.title('Sobreviventes menores de 15 anos por classe')

# Show the plot
plt.show()
```
