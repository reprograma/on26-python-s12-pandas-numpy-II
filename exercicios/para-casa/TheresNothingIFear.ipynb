```python
df = pd.read_csv("titanic.csv")
df
```

```python
# Gráfico númerico (1 variável)
# filtro com condicional
filter_hist = (df['Age'] > 0.8) & (df['Age'] < 70.72)

# gerando o histograma
df.loc[filter_hist, 'Age'].plot.hist(bins=2, edgecolor="White", color = "#ff6c00")

plt.xlabel("Classe", color = "Green") 
plt.ylabel("Quantidade", color = "Green") 
plt.title("Distribuição de Classes");
```

```python
df.loc[(df['Age'] == 80), :]
df.loc[filter_hist, 'Age'].plot.hist(bins=100, edgecolor="White", color = "#ff6c00")

plt.xlabel("Classe", color = "Green") 
plt.ylabel("Quantidade", color = "Green") 
plt.title("Distribuição de Classes");
```

```python
import random
# Gráficos de barras 3D
df.loc[(df['Age'] == 80), :]
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection="3d")
num_barras = 15
x_pos = random.sample(range(20), num_barras)
y_pos = random.sample(range(20), num_barras)
z_pos = [0] * num_barras
x_size = np.ones(num_barras)
y_size = np.ones(num_barras)
z_size = random.sample(range(20), num_barras)
plt.plot(x_pos, y_pos, z_pos, x_size, y_size, z_size, color='#ff6c00')
plt.title('Gráfico de barras 3D', fontsize=18);
plt.xlabel('Eixo X', fontsize=15)
plt.ylabel('Eixo Y', fontsize=15)

plt.show()
```
