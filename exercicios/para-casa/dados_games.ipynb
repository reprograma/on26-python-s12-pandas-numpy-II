## Processamento

Função de processamento


```python
import pandas as pd

# Carregar o arquivo CSV
df = pd.read_csv('Games.csv')
df

```

```python
# Traduzindo colunas para PT-BR
traducao_colunas = {
    "Name": "Nome",
    "Sales": "Vendas",
    "Series": "Série",
    "Release": "Lançamento",
    "Genre": "Gênero",
    "Developer": "Desenvolvedor",
    "Publisher": "Editora"
}

# Renomear as colunas com as traduções em português
df = df.rename(columns=traducao_colunas)
df
```

Leitura e tratamento do DF

```python
# dataframe originial
df
```

```python
# Removendo linhas com valores nulos
df = df.dropna()
df
```

Cenário 1: Filtro com uma condição

```python
# Filtro os jogos do gênero "First-person shooter"
df_filtrado = df[df['Gênero'] == 'First-person shooter']
df_filtrado.shape
```

Cenário 2: Filtro com múltiplas condições

```python
# Filtro os jogos do da Editora "Blizzard" e gênero "Action role-playing"
df_filtrado = df[(df['Editora'] == 'Blizzard Entertainment') & (df['Gênero'] == 'Action role-playing')]
df_filtrado.shape
```

Cenário 3: Seleção de dados com determinada string


```python
# Filtrando a linhas na coluna 'Editora' em que aparece a string 'Eletronic'
df_filtrado = df[df['Editora'].str.contains('Electronic')]
df_filtrado.shape
```

Cenário 4: Consulta utilizando Query


```python
# Para ilustrar essa função, vamos consultar jogos da Série The Witcher
df_consulta = df.query("Série == 'The Witcher'")
df_consulta
```

Cenário 5: Seleção de dados com loc

```python
# Selecionando apenas os nomes e as vendas dos jogos da série "The Sims"
df.loc[(df["Série"] == "The Sims"), ["Nome", "Vendas"]]
```

Cenário 6: Seleção de dados com iloc

```python
# Selecionando os três primeiros jogos do dataframe
df.iloc[:3, :]
```

Cenário 7: Slicing para um intervalo específico de linhas

```python
# Apenas as linhas de 10-20 no DataFrame
df[10:21]
```

Cenário 8: Agrupamento por gênero e soma de vendas

```python
# Agrupamento dos jogos por gênero e cálculo da soma das vendas para cada gênero
df.groupby("Gênero")["Vendas"].sum()
```

```python
# Agrupar os jogos por gênero e calcular a média das vendas apenas para o gênero "Ação"
media_vendas_acao = df[df["Gênero"] == "Action"]["Vendas"].mean()

print(f"A média das vendas do gênero 'Ação' é: {media_vendas_acao}")
```

```python
df[['Gênero']].describe(percentiles=[0.99]).T
```

```python
# Agrupar os jogos por gênero e encontrar os valores mínimos e máximos das vendas
min_vendas = df.groupby("Gênero")["Vendas"].min()
max_vendas = df.groupby("Gênero")["Vendas"].max()

print("Valores Mínimos de Vendas por Gênero:")
print(min_vendas)

print("\nValores Máximos de Vendas por Gênero:")
print(max_vendas)
```

# Visualizações / Insights
    

 - Código utilizado para geração de gráficos
   

```python
import matplotlib.pyplot as plt
```

- Gráficos

```python
# Gráfico de Barras para as Vendas por Gênero:
vendas_por_genero = df.groupby("Gênero")["Vendas"].sum()


# Criando um gráfico de barras
plt.figure(figsize=(12, 6))
vendas_por_genero.plot(kind="bar")
plt.title("Vendas por Gênero")
plt.xlabel("Gênero")
plt.ylabel("Vendas (em Milhões)")
plt.show()
```

 - Gráfico barras invertidas

```python

# Agrupando os dados por desenvolvedor e calculando as vendas totais para cada um
vendas_por_desenvolvedor = df.groupby("Desenvolvedor")["Vendas"].sum()

# Classificando em ordem descrescente os desenvolvedores
vendas_por_desenvolvedor = vendas_por_desenvolvedor.sort_values(ascending=False)

# Selecionando apenas os 10 principais desenvolvedores
top_10_desenvolvedores = vendas_por_desenvolvedor.head(10)

# Criando um gráfico de barras horizontais para as vendas por desenvolvedor (Top 10)
plt.figure(figsize=(10, 6))
top_10_desenvolvedores.plot(kind="barh")
plt.title("Vendas por Desenvolvedor (Top 10)")
plt.xlabel("Vendas (em Milhões)")
plt.ylabel("Desenvolvedor")
plt.show()
```

- Gráfico pizza

```python
# Agrupaando os dados por série e calculando as vendas totais para cada uma
vendas_por_serie = df.groupby("Série")["Vendas"].sum()

# Classificando as séries com base nas vendas em ordem decrescente
vendas_por_serie = vendas_por_serie.sort_values(ascending=False)

# Selecionando apenas as 5 séries mais populares
top_5_series = vendas_por_serie.head(5)

# Criando um gráfico de pizza para as principais séries de jogos
plt.figure(figsize=(8, 8))
plt.pie(top_5_series, labels=top_5_series.index, autopct="%1.1f%%")
plt.title("Principais Séries de Jogos")
plt.show()
```
