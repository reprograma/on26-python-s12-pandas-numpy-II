# Identificação dos dados

PassengerId: Este é um identificador único para cada passageiro no dataset. Ele não contém informações diretamente relevantes para a análise.

Survived: Esta coluna indica se um passageiro sobreviveu (1) ou não (0) ao naufrágio do Titanic. É uma informação crucial para entender a taxa de sobrevivência.

Pclass: Refere-se à classe do bilhete que o passageiro comprou. Existem três classes: 1ª, 2ª e 3ª. Esta informação é importante, pois geralmente reflete o status socioeconômico dos passageiros.

Name: O nome do passageiro. Embora contenha informações pessoais, ele não é usado diretamente para análises estatísticas.

Sex: Refere-se ao gênero do passageiro, que pode ser 'male' (masculino) ou 'female' (feminino).

Age: A idade do passageiro. Isso é importante para entender a distribuição etária dos passageiros e seu impacto na sobrevivência.

SibSp: Número de irmãos/cônjuges a bordo do Titanic com o passageiro.

Parch: Número de pais/filhos a bordo do Titanic com o passageiro.

Ticket: O número do bilhete do passageiro. Geralmente, não é relevante para análises estatísticas.

Fare: O valor pago pelo passageiro pelo bilhete. Pode ser útil para entender as diferenças nos preços dos bilhetes entre as classes e seu impacto na sobrevivência.

Cabin: O número da cabine do passageiro. Muitos valores estão faltando, então essa coluna pode não ser útil para análises.

Embarked: O porto de embarque do passageiro (C = Cherbourg, Q = Queenstown, S = Southampton). Pode ser relevante para entender se o porto de embarque afetou as chances de sobrevivência.

```python
# Importa a biblioteca Pandas
import pandas as pd

# Lê um arquivo CSV chamado "titanic.csv" e carrega seus dados em um DataFrame do Pandas
df = pd.read_csv("titanic.csv")

# Exibe o DataFrame (geralmente usado para verificar os dados)
df
```

```python
# Cria uma cópia do DataFrame original "df" e atribui a cópia a "df_backup"
df_backup = df.copy()

# Exibe "df_backup" (geralmente usado para verificar a cópia)
df_backup
```

```python
# Cria um dicionário de traduções das colunas do DataFrame
traducoes = {
    'PassengerId': 'IdPassageiro',
    'Survived': 'Sobreviveu', # 0 = Não, 1 = Sim
    'Pclass': 'Classe', # 1, 2, 3
    'Name': 'Nome',
    'Sex': 'Sexo',
    'Age': 'Idade',
    'SibSp': 'NumeroIrmaos',
    'Parch':  'NumeroPais',
    'Ticket': 'NumeroTicket',
    'Fare': 'PrecoTicket',
    'Cabin' : 'NumeroCabine',
    'Embarked': 'PortoEmbarcacao' # C = Cherbourg, Q = Queenstown, S = Southampton
}

# Cria uma lista vazia chamada "novas_colunas"
novas_colunas = []

# Itera sobre o dicionário de traduções e adiciona os valores (nomes traduzidos) à lista "novas_colunas"
for chave, valor in traducoes.items():
    novas_colunas.append(valor)

# Imprime a lista de novas colunas
print(novas_colunas)
```

```python
df.columns = novas_colunas
df
```

# Filtros


```python
# Filtra o DataFrame "df" com base em duas condições:
# 1. Seleciona as linhas onde a coluna "Idade" é maior que 30.
# 2. E, ao mesmo tempo, a coluna "Classe" é igual a 3.
df_fil = df[(df['Idade'] > 30) & (df['Classe'] == 3)]

# A linha acima cria um novo DataFrame, "df_fil," que contém apenas as linhas que atendem a essas condições.

# Em seguida, a linha abaixo exibe a forma (número de linhas e colunas) do DataFrame resultante.
df_fil.shape
```

# Consultas


```python
# Usa o método "query()" para realizar uma consulta no DataFrame "df"
# A consulta seleciona as linhas em que a coluna "Idade" é maior que 30.
df_consulta = df.query("Idade > 30")

# A linha acima cria um novo DataFrame chamado "df_consulta" que contém as linhas que atendem à condição especificada na consulta.

# Em seguida, exibe as primeiras linhas do DataFrame resultante.
df_consulta.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# Agrupar passageiros por classe
df_grupo_classe = df.groupby('Classe')

# O código acima utiliza o método "groupby()" do Pandas para agrupar as linhas do DataFrame "df" com base na coluna "Classe". 
# Isso cria grupos de dados separados para cada classe (1ª, 2ª e 3ª classe).

# Calcula a média de sobreviventes em cada classe
df_plot = df_grupo_classe['Sobreviveu'].mean()

# O código acima calcula a média da coluna "Sobreviveu" dentro de cada grupo de classe. 
# Isso fornece a taxa média de sobreviventes em cada classe.

# Cria um gráfico de barras
df_plot.plot.bar(color="pink")

# O código acima cria um gráfico de barras usando o Pandas. O eixo x do gráfico representa as classes, 
# o eixo y representa a média de sobreviventes e a cor das barras é definida como rosa ("pink").

# Configurações de rótulos e título
plt.xlabel('Classe')
plt.ylabel('Sobreviveu')
plt.title('Quantidade de sobreviventes por Classe')

# O código acima define rótulos para os eixos x e y do gráfico e atribui um título ao gráfico.

# Configuração de cores dos rótulos nos eixos x e y
plt.tick_params(axis='x', colors='purple')
plt.tick_params(axis='y', colors='purple')

# O código acima define a cor dos rótulos nos eixos x e y do gráfico como roxo ("purple").
```

```python
# Agrupar passageiros por gênero
df_grupo_genero = df.groupby('Sexo')

# O código acima utiliza o método "groupby()" do Pandas para agrupar as linhas do DataFrame "df" com base na coluna "Sexo". 
# Isso cria grupos de dados separados para passageiros do sexo masculino e feminino.

# Calcula a média de sobreviventes em cada gênero
df_plot = df_grupo_genero['Sobreviveu'].mean()

# O código acima calcula a média da coluna "Sobreviveu" dentro de cada grupo de gênero. 
# Isso fornece a taxa média de sobreviventes em cada gênero.

# Cria um gráfico de pizza
df_plot.plot(kind='pie')

# O código acima cria um gráfico de pizza usando o Pandas. O gráfico de pizza é uma forma de visualização que mostra as proporções das médias de sobreviventes em cada gênero.

# Define o título do gráfico
plt.title('Quantidade de sobreviventes por Gênero')

# O código acima atribui um título ao gráfico.
```

```python
# Agrupar passageiros pelo número de irmãos
df_grupo_irmaos = df.groupby('NumeroIrmaos')

# O código acima utiliza o método "groupby()" do Pandas para agrupar as linhas do DataFrame "df" com base na coluna "NumeroIrmaos". 
# Isso cria grupos de dados separados para diferentes números de irmãos a bordo.

# Calcula a média de sobreviventes em cada grupo
df_plot = df_grupo_irmaos['Sobreviveu'].mean()

# O código acima calcula a média da coluna "Sobreviveu" dentro de cada grupo de número de irmãos. 
# Isso fornece a taxa média de sobreviventes em relação ao número de irmãos.

# Cria um gráfico de linha
df_plot.plot(color='purple')

# O código acima cria um gráfico de linha usando o Pandas. O eixo x do gráfico representa o número de irmãos e o eixo y representa a média de sobreviventes. 
# A cor das linhas do gráfico é definida como roxa ("purple").

# Configurações de rótulos e título
plt.xlabel('Número de Irmãos')
plt.ylabel('Sobreviveu')
plt.title('Quantidade de sobreviventes com base no número de irmãos')

# O código acima define rótulos para os eixos x e y do gráfico e atribui um título ao gráfico.

# Configuração de cores dos rótulos nos eixos x e y
plt.tick_params(axis='x', colors='purple')
plt.tick_params(axis='y', colors='purple')

# O código acima define a cor dos rótulos nos eixos x e y do gráfico como roxa ("purple").
```
