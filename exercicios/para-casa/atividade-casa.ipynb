```python
import pandas as pd

def processamento_dados_csv(df):
    # Leitura do arquivo CSV
    df = pd.read_csv('dados.csv')
    
    # Exibir as primeiras 3 linhas
    df_primeiras_3_linhas = df.head(3)  
    
    #seleciona aleatoriamente 20 linhas do df
    df_menor = df.sample(20)

    # Arredondar para 2 casas decimais
    df_rounded = df.round(2)

    # Resumo do arredondamento
    resumo_arredondamento = df_rounded.describe().round(2)
    

    # Exibir colunas do DataFrame
    df_colunas = df.columns

    # Deletar coluna 'Altura'
    df = df.drop(columns = 'Altura')
   
    # Remover coluna 'UF'
    df = df.drop(columns = 'UF')    
    
    # ordenando os dados 
    df = df.sort_values("Idade")
    
    # Remover  dados duplicados
    df = df.drop_duplicates()
    
    # Resetar o índice
    df = df.reset_index(drop=True)
    
    
    #lista colunas 
    colunas = list(df.columns)
    
    
    # Verificar se 'Sexo' está presente antes de continuar
    if 'Sexo' in df.columns:
        # Selecionar apenas as colunas 'Iadade' 
        df.loc[0, 'Idade'] = 5

          # Checar valores nulos por coluna
        valores_nulos = df.isnull().sum()
        
        #metricas : analise descritiva
        metricas = df.describe(include= 'all')
    else:
        print("\nA coluna 'Sexo' não está presente no DataFrame.")
      

    return df_primeiras_3_linhas,df_menor, df_rounded, resumo_arredondamento, df_colunas, df, colunas, valores_nulos, metricas
```

```python
df_primeiras_3_linhas, df_menor, df_rounded, resumo_arredondamento, df_colunas, df_copia, colunas, valores_nulos, metricas = processamento_dados_csv(df)
```
