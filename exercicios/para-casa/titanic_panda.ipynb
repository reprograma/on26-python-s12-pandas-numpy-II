Introdução à Base de Dados do Titanic

A base de dados do Titanic é um conjunto de dados amplamente conhecido que contém informações sobre passageiros a bordo do navio RMS Titanic, que tristemente naufragou em sua viagem inaugural em 1912. Este conjunto de dados é frequentemente usado para análises exploratórias, visualizações de dados e modelagem preditiva.

Descrição da Base de Dados:

PassengerId: Um identificador único para cada passageiro.
Survived: Indica se o passageiro sobreviveu (1) ou não (0) ao desastre.
Pclass: A classe em que o passageiro viajava (1ª, 2ª ou 3ª classe).
Name: Nome do passageiro.
Sex: Gênero do passageiro.
Age: Idade do passageiro.
SibSp: Número de irmãos/cônjuges a bordo.
Parch: Número de pais/filhos a bordo.
Ticket: Número do bilhete.
Fare: Tarifa paga pelo passageiro.
Cabin: Número da cabine do passageiro.
Embarked: Porto de embarque do passageiro (C = Cherbourg, Q = Queenstown, S = Southampton).
Este conjunto de dados é uma fonte valiosa para entender os eventos que cercaram o naufrágio do Titanic e explorar como diferentes variáveis afetaram a sobrevivência dos passageiros.

```python
import matplotlib.pyplot as plt
df = pd.read_csv('titanic.csv')
```

```python

traducoes = {
    'PassengerId': 'IdPassageiro',
    'Survived': 'Sobreviveu', # 0 = Não, 1 = Sim
    'Pclass': 'Classe', # 1, 2, 3
    'Name': 'Nome',
    'Sex': 'Sexo',
    'Age': 'Idade',
    'SibSp': 'NumeroIrmaos',
    'Parch':  'NumeroPais',
    'Ticket': 'NumeroTicket',
    'Fare': 'PrecoTicket',
    'Cabin' : 'NumeroCabine',
    'Embarked': 'PortoEmbarcacao' # C = Cherbourg, Q = Queenstown, S = Southampton
}

# result = [(d['color'], d['value']) for d in dictionarylist]

novas_colunas = []
for chave, valor in traducoes.items():
    novas_colunas.append(valor)

print(novas_colunas)
```

```python
df.columns = novas_colunas
df
```

```python
df
```

```python
#Quantidade de passageiros por gênero

contagem_passageiros = df['Sexo'].value_counts()
contagem_passageiros.plot(kind='bar', color='gray', edgecolor='black')
plt.xlabel('Gênero', fontweight='bold', color='purple')
plt.title('Contagem de passageiros por gênero', color='purple')

plt.tick_params(axis='x', colors='red')
plt.tick_params(axis='y', colors='purple')

#conclusão havia mais homens que mulheres a bordo
```

```python
# Taxa de sobrevivência por faixa etária e gênero
faixas_etarias = ["Até 12 anos", "De 12 até 25 anos", "De 25 até 45 anos", "De 45 até 65 anos", "Acima de 65 anos"]
limites_idade = [0, 12, 25, 45, 65, float('inf')]

df['Faixa Etária'] = pd.cut(df['Idade'], bins=limites_idade, labels=faixas_etarias, right=False)


taxa_sob_faixa_etaria_sexo = df.groupby(['Faixa Etária', 'Sexo'])['Sobreviveu'].mean()

taxa_sob_faixa_etaria_sexo = taxa_sob_faixa_etaria_sexo.unstack()

taxa_sob_faixa_etaria_sexo.plot(kind='bar')
plt.xlabel('Faixa Etária', color='magenta')
plt.ylabel('Taxa de Sobrevivência', color='magenta')
plt.title('Taxa de Sobrevivência por Faixa Etária e Gênero', color='magenta')
plt.legend(title='Gênero', labels=['female', 'male'])
plt.show()

#conclusão: apesar de haverem mais homens a bordo, a taxa de sobrevivência foi mais alta no genêro feminino
```

```python
# gráfico de barras que conta o número de passageiros por classe e faixa etária
faixas_etaria = ["Até 12 anos", "De 12 até 25 anos", "De 25 até 45 anos", "De 45 até 65 anos", "Acima de 65 anos"]
limites_idade = [0, 12, 25, 45, 65, float('inf')]
df['Faixa Etária'] = pd.cut(df['Idade'], bins=limites_idade, labels=faixas_etaria, right=False)


contagem_passageiros = df.groupby(['Classe', 'Faixa Etária'])['IdPassageiro'].count().unstack()


contagem_passageiros.plot(kind='bar', stacked=True)
plt.xlabel('Classe', color='darkblue')
plt.ylabel('Número de Passageiros', color='darkblue')
plt.title('Número de Passageiros por Classe e Faixa Etária', color='darkblue')
plt.legend(title='Faixa Etária')
plt.show()

# aqui é possível visualizar que havia mais pessoas jovens na classe 3, enquanto na classe 1 havia pessoas com idades mais avançadas.
```
