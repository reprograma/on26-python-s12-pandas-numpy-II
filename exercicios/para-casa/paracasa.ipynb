Grafico 1 - Distribuição de idade dos passageiros 


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,8))
plt.hist(df['Idade'].dropna(), bins=10, color='red', edgecolor= 'yellow')
plt.title('Distribuição de Idade por Passageiros')
plt.xlabel('Idade')
plt.ylabel('Quantidade de Passageiros')
```

- De acordo com analise apresentada pelo o grafico. Conclui-se que a idade da maioria dos passageiros era entre 20 e 30 anos 

Grafico 2- Taxa de Sobrevivencia por Classe na Cabine 
- De inicio é necessario agrupar os dados por classe de cada cabine
- Calcular a taxa de sobrivência

```python
# Agrupar os dados por classe de cabine e calcular a taxa de sobrevivência
Classe_sobrevivente = df.groupby('Classe')['Sobreviveu'].mean()

# Código do gráfico 
plt.figure(figsize=(8, 6))
Classe_sobrevivente.plot(kind='bar', color='green')
plt.title('Taxa de Sobrevivência por Classe de Cabine')
plt.xlabel('Classe de Cabine')
plt.ylabel('Taxa de Sobrevivência')
plt.xticks(rotation=0)
plt.show()
```

De acordo com a analise em tela, estima-se que a classe 2,3 são as classes com menos sobrevivetes. E a classe 1 com mais sobrevivente.

Grafico 3 - Numero de passageiros por gênero
- Para essa analise foi apresentado o grafico em pizza


```python

contagem_por_genero = df['Sexo'].value_counts()

# Código do Gráfico
plt.figure(figsize=(3, 6))
plt.pie(contagem_por_genero, labels=contagem_por_genero.index, autopct='%1.1f%%', colors=['blue', 'red'])
plt.title('Contagem Por Gênero')
plt.show()
```

De acordo com a analise em tela, estima-se que o a maioria dos passageiros no navio titanic eram homens
