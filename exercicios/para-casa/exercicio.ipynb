```python
titanic_df = pd.read_csv('titanic.csv')
```

```python

titanic_df = processar_dados(titanic_df)
```

```python
plt.figure(figsize=(10, 5))
plt.hist(titanic_df['Age'].dropna(), bins=20, color='skyblue')
plt.xlabel('Idade')
plt.ylabel('Número de Passageiros')
plt.title('Distribuição de Idades dos Passageiros')
plt.show()
```

```python
survival_by_class = titanic_df.groupby('Pclass')['Survived'].mean()
survival_by_class.plot(kind='bar', color='lightcoral')
plt.xlabel('Classe')
plt.ylabel('Taxa de Sobrevivência')
plt.title('Taxa de Sobrevivência por Classe')
plt.xticks(rotation=0)
plt.show()
```

```python
survival_by_sex = titanic_df.groupby(['Sex', 'Survived']).size().unstack()
survival_by_sex.plot(kind='bar', stacked=True, color=['lightcoral', 'lightblue'])
plt.xlabel('Sexo')
plt.ylabel('Contagem de Passageiros')
plt.title('Distribuição de Sobreviventes e Não Sobreviventes por Sexo')
plt.xticks(rotation=0)
plt.legend(['Não Sobreviveu', 'Sobreviveu'])
plt.show()
```
