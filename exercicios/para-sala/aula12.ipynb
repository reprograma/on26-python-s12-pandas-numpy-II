```python
df_filtrado1 = df[(df["Idade"] < 11) & (df["Sexo"] == "male")]
df_filtrado2 = df[(df['Idade'] > 11) & (df['Idade'] < 20) & (df["Sexo"] == "male")]
```

```python
contagem_idades1 = df_filtrado1['Idade'].value_counts().sort_index()
contagem_idades2 = df_filtrado2['Idade'].value_counts().sort_index()
```

```python
contagem_idades1.plot(kind= "bar", edgecolor="maroon", color="pink")
plt.xlabel("Idade")
plt.ylabel("Contagem")
plt.title("Pessoas Masculinas Menores de 10 anos por Idade")
plt.show()
```

```python
contagem_idades2.plot(kind='bar', edgecolor= "black", color = "green")
plt.xlabel("Idade")
plt.ylabel("Contagem")
plt.title("Pessoas Masculinas Menores de 20 anos por Idade")
plt.show()
```

```python
df = df = pd.read_csv("C:\\Users\\nargy\\Documents\\estudos\\Semana-12\\on26-python-s12-pandas-numpy-II\\material\\titanic.csv")
```

```python
df_male = df[df["Sex"]=="male"]
```

```python
df_filtered = df_male[~df_male['Name'].str.contains('Mr.')]
contagem_palavras = df_filtered['Name'].str.split().explode().value_counts()
filtro = contagem_palavras.head(10)
```

```python
filtro.plot(kind='bar', edgecolor= "black", color = "purple")
plt.xlabel("Nome")
plt.ylabel("Contagem")
plt.title("Nomes mais comuns")
plt.show()
```
