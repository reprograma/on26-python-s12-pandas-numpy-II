```python
import pandas as pd

```

```python
# carregando o dataset 
df_dani = pd.read_csv("titanic.csv")
df_dani
```

```python
df_dani.columns

```

```python

traducoes = {
    'PassengerId': 'IdPassageiro',
    'Survived': 'Sobreviveu',
    'Pclass': 'Classe',
    'Name': 'Nome',
    'Sex': 'Sexo',
    'Age': 'Idade',
    'SibSp': 'IrmaosConjuge',
    'Parch': 'PaisFilhos',
    'Ticket': 'Bilhete',
    'Fare': 'Tarifa',
    'Cabin': 'Cabine',
    'Embarked': 'Embarque'
}

# Renomeie as colunas usando o dicionário de tradução
df_dani.rename(columns=traducoes, inplace=True)

# Agora as colunas estão traduzidas para português
print(df_dani.columns)
```

```python
df_dani
```

```python
#Filtrar quantos passageiros tinham menos de 18 anos
df_dani[(df_dani['Idade'] < 18)]
```

```python
#Determinar quantos adolescentes eram meninas 
df_dani_meninasadolescente = df_dani[(df_dani['Idade'] >= 12) & (df_dani['Idade'] <= 18 ) & (df_dani['Sexo'] == 'female')]
df_dani_meninasadolescente.reset_index(drop=True)
```

```python
#Passeigeiro com nome que tenha Dani
df_danistr = df_dani[df_dani['Nome'].str.contains('Dani')]
df_danistr.reset_index(drop=True)

```

```python
# Filtrando passageiros da segunda e terceira classe
df_daniclass = df_dani.query("Classe in (2, 3)")
df_daniclass.reset_index(drop=True)
```

```python
df_dani[:30]
```

```python
#determinar quantos passageiros sobreviveu
df_danilocv = df_dani.loc[df_dani["Sobreviveu"] == 1]
df_danilocv.reset_index(drop=True)

```

```python
nome_da_coluna = 'Idade'
df_danilocv = df_danilocv.sort_values(by=nome_da_coluna)
df_danilocv.reset_index(drop=True)
```

```python
df_danilocm = df_dani.loc[df_dani["Sobreviveu"] == 0]
df_danilocm.reset_index(drop=True)
```

```python
nome_da_coluna = 'Idade'
df_danilocm = df_danilocm.sort_values(by=nome_da_coluna)
df_danilocm.reset_index(drop=True)
```

```python
#Agrupar passageiros por Classe
dado_agrupados = df_dani.groupby('Classe')
dado_agrupados
```

```python
#contagem de Passageiros por classe
contagem_de_passegeiros = dado_agrupados['IdPassageiro'].count()
contagem_de_passegeiros
```

```python
#Função value counts
df_dani['Classe'].value_counts()
```

```python
#Media de Idade de passageiros por classe
media_idade = dado_agrupados['Idade'].mean()
media_idade
```

```python
df_dani[['Idade']].describe(percentiles=[1]).T
```

```python
#Obter o valor máximo e mínimo pago por passageiros em cada classe
valor_max = dado_agrupados['Tarifa'].max()
valor_min = dado_agrupados['Tarifa'].min()
print(f'Este é o valor máximo {valor_max}')
print(f'este o valor mínimo{valor_min}')
```

```python
#Agrupar passageiros por Sexo
dado_agrup_sex = df_dani.groupby('Sexo')
dado_agrup_sex
```

```python
#contagem de Passageiros por Sexo
contagem_de_passegeirosx = dado_agrup_sex['IdPassageiro'].count()
contagem_de_passegeirosx
```

```python
#Media de Idade de passageiros por Sexo
media_idadesx = dado_agrup_sex['Idade'].mean()
media_idadesx
```

```python
#Obter o valor máximo e mínimo pago por passageiros em cada classe
valor_maxsx = dado_agrup_sex['Tarifa'].max()
valor_minsx = dado_agrup_sex['Tarifa'].min()
print(f'Este é o valor máximo {valor_maxsx}')
print(f'este o valor mínimo{valor_minsx}')
```

```python
#Determinar valor máximo e mínimo pelo sexo

df_dani.groupby(['Sexo','Embarque']).agg({'Tarifa':['count','mean','min','max'],
 'Idade':['count','mean','min','max']})
```
